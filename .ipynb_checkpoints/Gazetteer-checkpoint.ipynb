```python
from bs4 import BeautifulSoup
import requests
import re

wikipedia_domain="https://en.wikipedia.org" 
entities=[]
links=[]

wikipedia_train_texts = []
wikipedia_test_texts = []



def crawl_text(wikipedia_link):    
    
    result  = requests.get(wikipedia_link)
    soup = BeautifulSoup(result.content)

    #delete script tags
    for x in soup.findAll('script'):
        x.extract()
    
    #get all paragraph element
    p_el = soup.find_all('p')
    all_text=""
    for x in p_el:    
        all_text=all_text+ " " + x.get_text()
                
    #find all entities and links
    for x in p_el:
        a_el=x.find_all('a', {'href': re.compile(r'^/wiki/')})    
        if len(a_el) > 0:
            for a in a_el:
                entity=a.get_text()
                link=wikipedia_domain+a['href']            
                
                entities.append(entity)
                links.append(link)
                
    return all_text



for x in train_link_array:
    wikipedia_train_texts.append(crawl_text(x))

print(len(wikipedia_train_texts))    
    
for x in test_link_array:
    wikipedia_test_texts.append(crawl_text(x))    

print(len(wikipedia_test_texts))


    
link_seed = pd.DataFrame(data={'entity': entities, 'link': links})      
link_seed                
```

```python
import re
import urllib

def get_header(x):
    text = re.search('(?<=https:\/\/en.wikipedia.org\/wiki\/).*', x).group()
    text = urllib.parse.unquote(urllib.parse.unquote(text))
    text = text.replace('_', ' ')
    
    if "#" in text: 
        text = text.split('#')[0]
    
    return text
    

link_seed['header'] = link_seed['link'].apply(lambda x:get_header(x))
link_seed
```

```python
link_seed_header = link_seed[['header','link']]
link_seed_header = link_seed_header.rename(index=str, columns={"header": "entity"})
link_seed_header

link_seed = link_seed[['entity','link']].append(link_seed_header)
link_seed
```

```python
def get_class(x):
    
    #print(x)
    
    class_type = "MISC"
    
    text = re.search('(?<=https:\/\/en.wikipedia.org\/wiki\/).*', x).group()
    text = urllib.parse.unquote(urllib.parse.unquote(text))
    
    uri= "http://lookup.dbpedia.org/api/search/PrefixSearch?QueryClass=&MaxHits=1&QueryString=" + text
    res = requests.get(uri)
    
    if not res.content:
        return class_type
    
    data = xmltodict.parse(res.content)
    
    obj = data['ArrayOfResult']    
    
    if "Result" in obj:    
        if type(obj["Result"]) == OrderedDict:
            res_dbpedia = obj["Result"]        
            uri_dbpedia = res_dbpedia["URI"]        
            text_dbpedia = re.search('(?<=http:\/\/dbpedia.org\/resource\/).*', uri_dbpedia).group()
            text_dbpedia = urllib.parse.unquote(urllib.parse.unquote(text_dbpedia))

            if text == text_dbpedia:
                classes = res_dbpedia["Classes"]
                if classes is not None:                      
                    res_classes = classes["Class"]
                    
                    if type(classes["Class"]) == OrderedDict:                        

                        if res_classes["Label"] == "person":
                            class_type = "PER"                        

                        if res_classes["Label"] == "organisation" or res_classes["Label"] == "organization":
                            class_type = "ORG"                        

                        if res_classes["Label"] == "place":
                            class_type = "LOC"

                    else:

                        for z in res_classes:
                            if z["Label"] == "person":
                                class_type = "PER"
                                break

                            if z["Label"] == "organisation" or z["Label"] == "organization":
                                class_type = "ORG"
                                break

                            if z["Label"] == "place":
                                class_type = "LOC"
                                break 
    
    #print(class_type)
    
    return class_type
    

#seed['class'] = seed['link'].apply(lambda x:get_class(x))
#seed

link_and_class['class'] = link_and_class['link'].apply(lambda x:get_class(x))
link_and_class
```

```python
def label_text(original_text, tokenized_text):
    
    patterns = []
    matched_words = []
    matched_words_class = []

    entities_class_array = seed_entity_class.values

    for x in entities_class_array: 
        W = x[0]
        C = x[1]
        words_found = set(re.findall(W, original_text))
        if len(words_found) > 0:
            for word in words_found:  
                matched_words.append(' '.join(nltk.word_tokenize(word.lower())))
                matched_words_class.append(C)

    #matched_words = list(set(matched_words))

    lookahead=0    
    for word in matched_words:
        nwords = word.count(' ')
        if nwords > lookahead:
            lookahead = nwords    
            

    i = 0
    l = len(tokenized_text)
    iobs = []

    while i < l:
        word = tokenized_text[i]
        j = i + 1  
        k = j + lookahead
        nextwords = []
        add_tag = False
        inside = False

        while j <= k:
            words = ' '.join([word] + nextwords)  
            words = words.lower()        
            #print(words)

            if words in matched_words:
                index = matched_words.index(words)
                if not inside:
                    iobs.append((word, 'B-' + matched_words_class[index]))

                for nword in nextwords:
                    iobs.append((nword, 'I-' + matched_words_class[index]))

                add_tag, inside = True, True
                i = j
                #break

            if j < l:
                nextword = tokenized_text[j]
                nextwords.append(nextword)
                j += 1
            else:
                break

        if not add_tag:
            inside = False
            i += 1
            iobs.append((word, 'O'))

    return iobs            


labeled_train_set = []
labeled_test_set = []

for x, y in zip(wikipedia_train_texts, tokenized_text_train):
    labeled_train_set.append(label_text(x, y))

print(len(labeled_train_set))    
    
for x, y in zip(wikipedia_test_texts, tokenized_text_test):
    labeled_test_set.append(label_text(x, y))    

print(len(labeled_test_set))
```

- save each labeled training set
- save labeled test set

```python
def save_to_text_file(file_name, tokens):
    with open(file_name, "w") as text_file:    
        for wrd, lbl in tokens:            
            text_file.write(wrd + ' ' + lbl + '\n')    


train_file_idx = 0
test_file_idx = 0

for x in labeled_train_set:
    file_name = "train-data/train-" + str(train_file_idx) + ".txt"
    
    save_to_text_file(file_name, x)
    
    train_file_idx += 1

    
    
for x in labeled_test_set:
    file_name = "test-data/test-" + str(test_file_idx) + ".txt"
    
    save_to_text_file(file_name, x)
    
    test_file_idx += 1    


    
    
print("done")    
```
