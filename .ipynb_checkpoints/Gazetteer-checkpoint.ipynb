```python
import pandas as pd
import numpy as np

wikidata = pd.read_csv('wikidata.csv')
```

```python
wikidata = wikidata.dropna()
wikidata
```

```python
seed = wikidata[['engine','article']]
```

```python
seed_alias = wikidata[['alias','article']]
seed_alias = seed_alias.rename(index=str, columns={"alias": "engine"})
seed_alias

seed = seed.append(seed_alias)
seed
```

```python
test_link = seed.article.unique()[0]
test_link
```

```python
from bs4 import BeautifulSoup
import requests
import re

result  = requests.get(test_link)
soup = BeautifulSoup(result.content)

#delete script tags
for x in soup.findAll('script'):
    x.extract()
    
#get all paragraph element
p_el = soup.find_all('p')
all_text=""
for x in p_el:    
    all_text=all_text+ " " + x.get_text()

#find all entities and links
wikipedia_domain="https://en.wikipedia.org" 
entities=[]
links=[]
for x in p_el:
    a_el=x.find_all('a', {'href': re.compile(r'^/wiki/')})    
    if len(a_el) > 0:
        for a in a_el:
            entity=a.get_text()
            link=wikipedia_domain+a['href']            
            entities = np.append(entities, entity)
            links = np.append(links, link)

link_seed = pd.DataFrame(data={'entity': entities, 'link': links})      
link_seed
```

```python
patterns = []
matched_words = []

for x in seed_entity[['entity']].values.flatten(): 
    words_found = set(re.findall(x, all_text))
    if len(words_found) > 0:
        for word in words_found:  
            matched_words.append(word)         

matched_words = list(set(matched_words))
            
lookahead=0    
for word in matched_words:
    nwords = word.count(' ')
    if nwords > lookahead:
        lookahead = nwords
```

```python
import nltk

tokenized_text = nltk.word_tokenize(all_text)
```

```python
i = 0
l = len(tokenized_text)
iobs = []

while i < l:
    word = tokenized_text[i]
    j = i + 1  
    k = j + lookahead
    nextwords = []
    add_tag = False
    inside = False

    while j <= k:
        words = ' '.join([word] + nextwords)  
                
        #print(words)
        
        if words in matched_words:
            index = matched_words.index(words)
            if not inside:
                iobs.append((word, 'B'))

            for nword in nextwords:
                iobs.append((nword, 'I'))

            add_tag, inside = True, True
            i = j
            #break
            
        if j < l:
            nextword = tokenized_text[j]
            nextwords.append(nextword)
            j += 1
        else:
            break

    if not add_tag:
        inside = False
        i += 1
        iobs.append((word, 'O'))

iobs
```

```python
import xmltodict

result  = requests.get("http://lookup.dbpedia.org/api/search/PrefixSearch?QueryClass=&MaxHits=5&QueryString=W%C3%A4rtsil%C3%A4")
data = xmltodict.parse(result.content)
```

```python
from collections import OrderedDict

obj = data['ArrayOfResult']
results = obj["Result"] if type(obj) == OrderedDict else [obj["Result"]]

for x in results:
    print(x["Label"])
    print(x["URI"])
    
    if x["Classes"] is not None:
        print(x["Classes"])
        
        classes = x["Classes"]
        res_classes = classes["Class"] if type(obj) == OrderedDict else [classes["Class"]]                
        for y in res_classes:            
            if y["Label"] == "person":                
                print("PER")
                break
                
            if y["Label"] == "organisation" or y["Label"] == "organization":
                print("ORG")
                break
                
            if y["Label"] == "place":
                print("LOC")
                break
    
    print("\n")
```

```python
def get_class(x):
    
    print(x)
    
    class_type = "MISC"
    
    text = re.search('(?<=https:\/\/en.wikipedia.org\/wiki\/).*', x).group()
    text = urllib.parse.unquote(urllib.parse.unquote(text))
    
    uri= "http://lookup.dbpedia.org/api/search/PrefixSearch?QueryClass=&MaxHits=1&QueryString=" + text
    res = requests.get(uri)
    data = xmltodict.parse(res.content)
    
    obj = data['ArrayOfResult']    
    
    if "Result" in obj:    
        if type(obj["Result"]) == OrderedDict:
            res_dbpedia = obj["Result"]        
            uri_dbpedia = res_dbpedia["URI"]        
            text_dbpedia = re.search('(?<=http:\/\/dbpedia.org\/resource\/).*', uri_dbpedia).group()
            text_dbpedia = urllib.parse.unquote(urllib.parse.unquote(text_dbpedia))

            if text == text_dbpedia:
                classes = res_dbpedia["Classes"]
                if classes is not None:                      
                    res_classes = classes["Class"]
                    
                    if type(classes["Class"]) == OrderedDict:                        

                        if res_classes["Label"] == "person":
                            class_type = "PER"                        

                        if res_classes["Label"] == "organisation" or res_classes["Label"] == "organization":
                            class_type = "ORG"                        

                        if res_classes["Label"] == "place":
                            class_type = "LOC"

                    else:

                        for z in res_classes:
                            if z["Label"] == "person":
                                class_type = "PER"
                                break

                            if z["Label"] == "organisation" or z["Label"] == "organization":
                                class_type = "ORG"
                                break

                            if z["Label"] == "place":
                                class_type = "LOC"
                                break 
    
    print(class_type)
    
    return class_type
    

seed['class'] = seed['link'].apply(lambda x:get_class(x))
seed
```
